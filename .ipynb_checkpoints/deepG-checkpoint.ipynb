## Lower bound

### Objective function (equation (3))

```python
# Objective functions
def f(variables):
    '''Objective function with variables = [wl, bl]'''
    #return df.apply(lambda x: x["pixel_value"]-x["rotation_value"]*variables[0]+variables[1], axis=1).sum()
    return inverse_nb_values*(sum_pixel_values - sum_rotation_values*variables[0] - nb_values*variables[1])


```

### Constraints (equation (4))

```python
# Constraints 
constraint1 = df.apply(lambda x: [x["rotation_value"], 1], axis=1).tolist()
constraint_bound_inf = [-np.inf]*df.shape[0]
constraint_bound_sup = df["pixel_value"].tolist()

linear_constraint = LinearConstraint(
    constraint1 , constraint_bound_inf, constraint_bound_sup)
```

```python
constraint_bound_sup
```

```python
# Starting point wl=-0.9, bl=1.7
variables_start = np.array([-0.2,1.5])

# Optimization
result = minimize(
    f, variables_start, method='trust-constr', constraints=linear_constraint)

results = result.x.round(2)
results
```

## Upper bound

```python
# Objective functions
def f_upper(variables):
    '''Objective function with variables = [wl, bl]'''
    #return df.apply(lambda x: x["pixel_value"]-x["rotation_value"]*variables[0]+variables[1], axis=1).sum()
    return -inverse_nb_values*(sum_pixel_values - sum_rotation_values*variables[0] - nb_values*variables[1])
```

```python
linear_constraint_upper = LinearConstraint(
    constraint1 , constraint_bound_sup, [np.inf]*df.shape[0])
# Optimization
result_upper = minimize(
    f_upper, variables_start, method='trust-constr', constraints=linear_constraint_upper)

results_upper = result_upper.x.round(2)
results_upper
```

```python
# Let's define arbitrary the lower bound wlr and upper bound wur for the rotation
wlr, wur = -0.9, -0.5

# Let's define arbitrary the lower bound wlb and upper bound wub for the brightness
wlb, wub = -0.4, -0.3

# Let's define arbitrary the lower bound lbias and upper bound ubias for the bias
lbias, ubias = 1.07, 1.07

def visualization(wlr, wur, wlb, wub, lbias, ubias):

    # The range of values are determined by the user 
    rotation = np.linspace(0, np.pi/4*180/np.pi, 100)
    brightness = np.linspace(0, 0.00008, 100)

    # create a rectangular grid 
    Rotation, Brightness = np.meshgrid(rotation, brightness)



    print(wlr)

    # Calculate pixel values for each hyperplane
    Ik_lower = wlr * Rotation + wlb * Brightness + lbias
    Ik_upper = wur * Rotation + wub * Brightness + ubias


    fig = go.Figure(data=[
        # Lower bound hyperplane pixel value
        go.Surface(z=Ik_lower, x=Brightness, y=Rotation, colorscale=[[0, 'orange'], [1, 'pink']], showscale=False),
        # Upper bound hyperplane pixel value
        go.Surface(z=Ik_upper, x=Brightness, y=Rotation, colorscale=[[0, 'orange'], [1, 'orange']], showscale=False, opacity=0.5)
    ])
    
    fig.add_trace(
    go.Scatter3d(x=[0, 0.00002],
                 y=[0.1],
                 z=[0.98],
                 mode='markers')
    )

    fig.add_trace(
    go.Scatter3d(x=[0, 0.00002],
                 y=[0.2],
                 z=[0.97],
                 mode='markers')
    )

    fig.add_trace(
    go.Scatter3d(x=[0, 0.00002],
                 y=[0.5],
                 z=[0.79],
                 mode='markers')
    )
    
    fig.add_trace(
    go.Scatter3d(x=[0, 0.00002],
                 y=[0.7],
                 z=[0.44],
                 mode='markers')
    )

    fig.update_layout(
        scene=dict(
            xaxis_title='Brightness',
            yaxis_title='Rotation (rad)',
            zaxis_title='Pixel Value',
            yaxis=dict(range=[0, np.pi/4]),
            xaxis=dict(range=[0, 0.00008], autorange='reversed'),  # Reversing the x-axis
            zaxis=dict(range=[0, 1]),
            aspectmode='cube'
        ),
        title='Hyperplanes in Pixel Value Space'
    )
    
    # Show the plot
    fig.show()

visualization(wlr, wur, wlb, wub, lbias, ubias)
```
